# Dimensionality Reduction on 'Blobs' Dataset

This notebook applies 9 different dimensionality reduction techniques to the synthetic 'blobs' dataset.

**Dataset:** 300 samples from 3 Gaussian clusters.

**Methods included:**
1. PCA
2. MDS
3. Isomap
4. Locally Linear Embedding (LLE)
5. Laplacian Eigenmap
6. Diffusion Map *(run locally)*
7. t-SNE
8. UMAP
9. Autoencoder

```python
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generate blobs dataset
X, y = make_blobs(n_samples=300, centers=3, random_state=42)
X = StandardScaler().fit_transform(X)

sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, palette='Set2')
plt.title('Original Data: Blobs')
plt.show()
```

## PCA

```python
from sklearn.decomposition import PCA
X_pca = PCA(n_components=2).fit_transform(X)
```

```python
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, palette='Set2')
plt.title('PCA Embedding')
plt.show()
```

## MDS

```python
from sklearn.manifold import MDS
X_mds = MDS(n_components=2, random_state=42).fit_transform(X)
```

```python
sns.scatterplot(x=X_mds[:, 0], y=X_mds[:, 1], hue=y, palette='Set2')
plt.title('MDS Embedding')
plt.show()
```

## Isomap

```python
from sklearn.manifold import Isomap
X_iso = Isomap(n_components=2).fit_transform(X)
```

```python
sns.scatterplot(x=X_iso[:, 0], y=X_iso[:, 1], hue=y, palette='Set2')
plt.title('Isomap Embedding')
plt.show()
```

## Laplacian Eigenmap

```python
from sklearn.manifold import SpectralEmbedding
X_lap = SpectralEmbedding(n_components=2, random_state=42).fit_transform(X)
```

```python
sns.scatterplot(x=X_lap[:, 0], y=X_lap[:, 1], hue=y, palette='Set2')
plt.title('Laplacian Eigenmap Embedding')
plt.show()
```

*Run this cell locally after installing `pydiffmap`.*

## t-SNE

```python
from sklearn.manifold import TSNE
X_tsne = TSNE(n_components=2, random_state=42).fit_transform(X)
```

```python
sns.scatterplot(x=X_tsne[:, 0], y=X_tsne[:, 1], hue=y, palette='Set2')
plt.title('t-SNE Embedding')
plt.show()
```

## UMAP

```python
import umap
X_umap = umap.UMAP(n_components=2, random_state=42).fit_transform(X)
```

```python
sns.scatterplot(x=X_umap[:, 0], y=X_umap[:, 1], hue=y, palette='Set2')
plt.title('UMAP Embedding')
plt.show()
```

## Autoencoder

```python
from sklearn.neural_network import MLPRegressor
ae = MLPRegressor(hidden_layer_sizes=(5,), max_iter=1000, random_state=42)
X_scaled = StandardScaler().fit_transform(X)
ae.fit(X_scaled, X_scaled)
X_auto = ae.predict(X_scaled)[:, :2]
```

```python
sns.scatterplot(x=X_auto[:, 0], y=X_auto[:, 1], hue=y, palette='Set2')
plt.title('Autoencoder Embedding')
plt.show()
```

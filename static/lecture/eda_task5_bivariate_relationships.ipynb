# EDA Task 5: Bivariate Relationships

## 📘 What Is a Bivariate Relationship?
A bivariate relationship examines how **two variables** relate to each other — whether they change together, differ across groups, or predict one another.

---

## 🎯 Why This Matters in EDA
- Identify potential **predictor–response** patterns
- Understand **differences across groups**
- Guide **modeling** and **hypothesis formation**

---

## 🧰 Tools for Bivariate Analysis

| Variable Types              | Tool/Plot                             | Use Case                                      |
|-----------------------------|----------------------------------------|-----------------------------------------------|
| Numeric vs. Numeric         | `scatterplot`, `jointplot`, `lmplot`, `corr()` | Relationship trends, correlation strength     |
| Categorical vs. Numeric     | `boxplot`, `violinplot`, `stripplot`, `swarmplot` | Compare distributions across groups          |
| Categorical vs. Categorical | `crosstab`, `heatmap`, grouped bars`  | Frequencies and proportions between categories |


## 📁 Activity 1: Titanic Dataset — Bivariate Analysis

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
df_titanic = sns.load_dataset('titanic')

# Scatterplot: Age vs. Fare
sns.scatterplot(x='age', y='fare', data=df_titanic)
plt.title('Scatterplot: Age vs. Fare')
plt.show()

# Boxplot: Fare by Class
sns.boxplot(x='class', y='fare', data=df_titanic)
plt.title('Fare by Class')
plt.show()

# Violin Plot: Age by Sex
sns.violinplot(x='sex', y='age', data=df_titanic)
plt.title('Age by Sex')
plt.show()

# Crosstab heatmap: Class vs. Embarked
ct = pd.crosstab(df_titanic['class'], df_titanic['embarked'], normalize='index')
sns.heatmap(ct, annot=True, cmap='Blues')
plt.title('Proportion of Embarked Ports by Class')
plt.show()
```

## 🔬 Activity 2: Simulated Dataset — Bivariate Analysis

```
import numpy as np

# Simulate dataset
np.random.seed(0)
df_sim = pd.DataFrame({
    'age': np.random.randint(18, 80, 100),
    'income': np.random.normal(50000, 15000, 100),
    'gender': np.random.choice(['male', 'female'], size=100),
    'purchased': np.random.choice([0, 1], size=100)
})

# Scatterplot: Age vs. Income
sns.scatterplot(x='age', y='income', data=df_sim)
plt.title('Scatterplot: Age vs. Income')
plt.show()

# Boxplot: Income by Gender
sns.boxplot(x='gender', y='income', data=df_sim)
plt.title('Income by Gender')
plt.show()

# Crosstab heatmap: Gender vs. Purchased
ct_sim = pd.crosstab(df_sim['gender'], df_sim['purchased'], normalize='index')
sns.heatmap(ct_sim, annot=True, cmap='Purples')
plt.title('Purchase Outcome by Gender')
plt.show()
```

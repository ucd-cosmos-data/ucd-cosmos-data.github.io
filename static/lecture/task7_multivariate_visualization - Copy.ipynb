# Task 7: Classical Multivariate Visualization

This notebook showcases core multivariate visualization tools (excluding PCA/t-SNE) using both the Titanic dataset and a simulated dataset.

Covered methods:
1. Enhanced Scatterplot (hue, size, style)
2. Pairplot with hue
3. FacetGrid and Catplot
4. Parallel Coordinates
5. (Optional) 3D Scatterplot

## 1. Enhanced Scatterplot (Titanic)
- Variables: `age` (x), `fare` (y), `class` (hue), `survived` (style)

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df_titanic = sns.load_dataset('titanic')
df_titanic = df_titanic.dropna(subset=['age', 'fare', 'class', 'survived'])

sns.scatterplot(data=df_titanic, x='age', y='fare', hue='class', style='survived')
plt.title('Titanic: Age vs Fare by Class and Survival')
plt.show()
```

## 2. Pairplot with Hue (Titanic)
- Shows relationships among multiple numeric variables grouped by survival

```python
sns.pairplot(df_titanic[['age', 'fare', 'pclass', 'sibsp', 'parch', 'survived']], hue='survived')
plt.suptitle('Titanic: Pairplot with Hue = Survived', y=1.02)
plt.show()
```

## 3. FacetGrid and Catplot (Titanic)
- Plot survival rate by class, split by sex

```python
sns.catplot(data=df_titanic, x='class', hue='survived', col='sex', kind='count')
plt.suptitle('Titanic: Survival by Class, Faceted by Sex', y=1.05)
plt.show()
```

## 4. Parallel Coordinates (Simulated Data)
- Compare multi-feature profiles by group

```python
from pandas.plotting import parallel_coordinates
import numpy as np

# Simulate dataset
np.random.seed(0)
n = 150
sim_df = pd.DataFrame({
    'age': np.random.randint(20, 70, size=n),
    'income': np.random.normal(50000, 10000, size=n),
    'spend_score': np.random.normal(70, 10, size=n),
    'visits': np.random.poisson(4, size=n),
    'group': np.random.choice(['A', 'B', 'C'], size=n)
})

parallel_coordinates(sim_df, class_column='group', color=('#1f77b4', '#ff7f0e', '#2ca02c'))
plt.title('Simulated: Parallel Coordinates by Group')
plt.show()
```

# Task 8: Dimensionality Reduction

This notebook applies 8 dimensionality reduction techniques to 4 datasets.

**Methods included:**
- PCA
- MDS
- Isomap
- Locally Linear Embedding (LLE)
- Laplacian Eigenmaps
- t-SNE
- UMAP
- Autoencoder

❗ *Diffusion Maps not included here due to environment limitations. See separate instructions to run locally.*

```python
%pip install umap-learn

from sklearn.datasets import make_moons, make_circles, make_blobs, make_classification
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import MDS, Isomap, TSNE, LocallyLinearEmbedding, SpectralEmbedding
from sklearn.neural_network import MLPRegressor
import umap
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Create datasets
datasets = {}
datasets['blobs'] = make_blobs(n_samples=300, centers=3, random_state=42)
datasets['moons'] = make_moons(n_samples=300, noise=0.05, random_state=42)
datasets['circles'] = make_circles(n_samples=300, noise=0.05, factor=0.5, random_state=42)
datasets['classification'] = make_classification(n_samples=300, n_features=10, n_informative=5,
    n_redundant=2, n_classes=3, n_clusters_per_class=1, random_state=42)
```

```python
methods = {
    'PCA': lambda X: PCA(n_components=2).fit_transform(X),
    'MDS': lambda X: MDS(n_components=2, random_state=42).fit_transform(X),
    'Isomap': lambda X: Isomap(n_components=2).fit_transform(X),
    'LLE': lambda X: LocallyLinearEmbedding(n_components=2, n_neighbors=12, random_state=42).fit_transform(X),
    'Laplacian': lambda X: SpectralEmbedding(n_components=2, random_state=42).fit_transform(X),
    't-SNE': lambda X: TSNE(n_components=2, learning_rate='auto', init='pca', random_state=42).fit_transform(X),
    'UMAP': lambda X: umap.UMAP(n_components=2, random_state=42).fit_transform(X),
    'Autoencoder': lambda X: MLPRegressor(hidden_layer_sizes=(5,), max_iter=1000, random_state=42)
        .fit(StandardScaler().fit_transform(X), StandardScaler().fit_transform(X))
        .predict(StandardScaler().fit_transform(X))[:, :2]
}

fig, axes = plt.subplots(len(datasets), len(methods), figsize=(24, 16))

for i, (ds_name, (X, y)) in enumerate(datasets.items()):
    X_scaled = StandardScaler().fit_transform(X)
    for j, (method_name, reducer) in enumerate(methods.items()):
        ax = axes[i, j]
        try:
            X_embedded = reducer(X_scaled)
            ax.scatter(X_embedded[:, 0], X_embedded[:, 1], c=y, cmap='Spectral', s=10)
            ax.set_title(f"{ds_name} - {method_name}")
        except Exception as e:
            ax.set_title(f"{ds_name} - {method_name}\n(Failed)")
            ax.axis('off')
        ax.set_xticks([])
        ax.set_yticks([])

plt.tight_layout()
plt.suptitle("2D Visualizations using Dimensionality Reduction", y=1.02, fontsize=16)
plt.show()
```

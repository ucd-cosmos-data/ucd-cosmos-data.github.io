# `pandas`

Basically, each Python library defines one or more **core** classes that represent its core functionality, along with additional utility functions that support or extend those core features. 

Let’s take a closer look at the `pandas` library from this point of view.

**Core classes**
- `Index`: Class that labels rows or columns
- `Series`: 1D labeled array
- `DataFrame`: 2D labeled array

More specifically, the class hierarchy in pandas can be summarized as follows:

```text
NDFrame: collection of data
 ├── Series: 1D array
 └── DataFrame: 2D array (two Index classes (row/column) and multiple Series classes)

 Index: labels used to identify rows or columns
 ├── RangeIndex
 ├── DatetimeIndex
 ├── MultiIndex
 └── ...
 ```

````python
import pandas as pd
````

### `DataFrame`

````python
sales = pd.DataFrame({
    "date": ["2024-01-05", "2024-01-07", "2024-01-10", "2024-02-02", "2024-02-05", "2024-02-10"],
    "region": ["East", "West", "East", "South", "West", "East"],
    "product": ["B", "A", "B", "B", "A", "B"],
    "units": [10, 6, 15, 6, 8, None],
    "price": [2.50, 3.00, 4.00, 4.00, None, 4.50],
    "returned": [False, False, True, False, False, True]
})
````

````python
print(type(sales))
````

````python
sales_idx = sales.index
sales_col = sales.columns
````

````python
sales_idx
````

````python
print(type(sales_idx))
````

````python
sales_col
````

````python
print(type(sales_col))
````

#### Indexing: selecting by label and position

- `.loc[row_labels, column_labels]` selects by index/column labels.
- `.iloc[row_positions, column_positions]` selects by integer positions.
- `.set_index()`, `.reset_index()`, and `.sort_index()` change or organize the index.

````python
sales.loc[0] # selecting a single row by index label
## Note that this returns a Series, not a DataFrame.
````

````python
sales.loc[[0]] # This returns a DataFrame, not a Series.
````

````python
sales.loc[:, "region"] # selecting a single column by label
## Note that this returns a Series, not a DataFrame.
````

````python
sales.loc[:, ["region"]] # This returns a DataFrame, not a Series.
````

````python
sales.loc[[0,1,2]] # selecting multiple rows by index label
````

````python
sales.loc[0:2]
````

````python
sales.loc[:, ["region", "units"]] # selecting multiple columns by label
````

````python
sales.loc[0:2, ["region", "units"]]
````

````python
sales.loc[[0,2,4], ["region", "units"]]
````

````python
sales.iloc[0:3, [0, 2, 3]]
````

#### Index
`set_index()` and `reset_index()`

````python
sales.set_index("date") 
# This returns a new DataFrame with the "date" column as the index, but does not modify the original DataFrame.
````

````python
sales_by_date = sales.set_index("date")
sales_by_date
````

````python
sales_by_date.index
````

````python
sales_by_date.loc["2024-01"] # Can I select rows by a partial date string (e.g. YYYY-MM)?
````

````python
# You can do that if the index is in datetime format.
sales_by_date.index = pd.to_datetime(sales_by_date.index) # Convert the index to datetime format
sales_by_date.index
````

````python
sales_by_date.loc["2024-01"] # selecting rows by index label (date range)
````

````python
sales_by_date.reset_index() # resetting the index. 
## Note that this method returns a new DataFrame and does not modify the original DataFrame in place.
````

````python
sales_by_date
````

#### Subsetting: filtering rows and choosing columns

Use column lists for columns, boolean conditions for rows.

````python
sales["region"] # selecting a single column by label
````

````python
sales[["region", "product", "units"]] # selecting multiple columns by label
````

````python
sales[sales["units"] >= 10] # filtering rows based on a condition
````

````python
mask = (sales["region"] == "East") & (sales["units"] >= 10) # creating a boolean mask for filtering rows
mask
````

````python
sales[mask] # filtering rows using the boolean mask
````

````python
sales.loc[mask, ["date", "region", "units"]] 
# filtering rows using the boolean mask and selecting specific columns
````

#### Sorting methods

`sort_values()` sorts all rows.

````python
sales.sort_values("units", ascending=False) # sort by one column
````

````python
sales.sort_values(["units", "price"], ascending=[True, False]) # sort by multiple columns
````

### Value counts
`value_counts()` returns eqch unique value and its count.

````python
sales["product"].value_counts() # count occurrences of each unique value in a column
````

````python
sales["region"].value_counts(ascending=False)
````

````python
sales[["region", "product"]].value_counts()
````

#### Column operations

Common column operations include creating new columns, deleting columns, renaming columns, and changing data types.

````python
sales["revenue"] = sales["units"] * sales["price"] # adding a new column with []
# This method directly modifies the DataFrame in place and does not return a new DataFrame.
````

````python
sales
````

````python
sales.drop(columns=["returned"]) # deleting columns
# This method returns a new DataFrame and does not modify the original DataFrame in place.
````

````python
sales.rename(columns={"region": "sales_region", "units": "units_sold"}) # renaming columns
# This method returns a new DataFrame and does not modify the original DataFrame in place.
````

````python
sales.astype({"region": "category", "product": "category", "units": "Int64"}) # changing data types of columns
# This method returns a new DataFrame and does not modify the original DataFrame in place.
````

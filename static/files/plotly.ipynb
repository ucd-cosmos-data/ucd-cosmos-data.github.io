# Plotly Express


`plotly` is a Python graphing library for creating interactive plots. In this notebook, we will focus on `plotly.express`, usually imported as `px`.

`plotly.express` is the high-level interface of Plotly. It is designed for quickly making common chart types from a tidy `pandas` DataFrame.


````python
import numpy as np
import pandas as pd
import plotly.express as px
````

A useful way to understand Plotly Express is to think about the general plotting workflow:

```python
# data prep
df = pd.DataFrame(...)

# create a figure with plotly.express
fig = px.DESIRED_PLOT(df, x=..., y=..., color=..., ...)

# customize the graphical marks
fig.update_traces(...)

# customize the overall figure layout
fig.update_layout(...)

# render the figure
fig.show()
```

So the common pattern is:

```python
fig = px.DESIRED_PLOT(...)
fig.update_traces(...)
fig.update_layout(...)
fig.show()
```

One good thing about `plotly` is that it is `pandas`-friendly.

## I. Figure, Traces, and Layout


Plotly plots are built around a few important ideas:

- The **figure** is the complete plot object.
- A figure contains one or more **traces**. A trace is a set of graphical marks, such as points, bars, lines, or boxes.
- The **layout** controls the overall appearance: title, axes, legend, size, margins, template, background, etc.
- `fig.show()` renders the interactive plot.

In Matplotlib, you usually think in terms of `figure` and `axes`. In Plotly, it is often more useful to think in terms of `figure`, `traces`, and `layout`.


````python
df = pd.DataFrame({
    "x": [1, 2, 3, 4, 5],
    "y": [4, 6, 5, 8, 7],
    "group": ["A", "A", "B", "B", "B"]
})

fig = px.scatter(df, x="x", y="y", color="group")

print(type(fig))
print("Number of traces:", len(fig.data))
print("Trace types:", [trace.type for trace in fig.data])

fig.show()
````

Because we used `color="group"`, Plotly Express created separate traces for the groups. This is common in Plotly: visual mappings such as `color`, `symbol`, `line_dash`, and `facet_col` can automatically create multiple traces or panels.


## II. Plotly Express Functions


#### II.1. Basic Plot Types

| Type | Method | Description |
| --- | --- | --- |
| Line | `px.line()` | Line plot |
| Scatter | `px.scatter()` | Scatter plot |
| Bar | `px.bar()` | Bar plot |
| Histogram | `px.histogram()` | Histogram |

| Violin | `px.violin()` | Violin plot |

| Area | `px.area()` | Area plot |
| Heatmap/Image | `px.imshow()` | Display a matrix or image |

Each method has its own arguments. You do not need to memorize all of them. Use the official documentation, examples, autocomplete, or an AI assistant when you need a specific option.


````python
df = pd.DataFrame({
    "day": ["Mon", "Tue", "Wed", "Thu", "Fri"],
    "sales": [12, 18, 15, 22, 26]
})

fig = px.bar(df, x="day", y="sales")
fig.show()
````

````python
df = pd.DataFrame({
    "hours_studied": [1, 2, 2.5, 3, 4, 5, 5.5, 6],
    "score": [62, 68, 70, 73, 80, 86, 88, 93],
    "section": ["A", "A", "A", "B", "B", "B", "B", "A"]
})

fig = px.scatter(df, x="hours_studied", y="score", color="section")
fig.show()
````

#### II.2. Mapping Columns to Visual Properties

Plotly Express works especially well with DataFrames. Most arguments refer to column names:

| Argument | Meaning |
| --- | --- |
| `x`, `y` | Columns used for the x and y axes |
| `color` | Column used to color groups |
| `size` | Column used to change marker size |
| `symbol` | Column used to change marker symbol |
| `hover_name`, `hover_data` | Columns shown in the hover tooltip |
| `text` | Column used for text labels on marks |

This is similar to saying: "Use this column for position, this column for color, this column for size, and so on."


````python
df = px.data.gapminder().query("year == 2007")

fig = px.scatter(
    df,
    x="gdpPercap",
    y="lifeExp",
    color="continent",
    size="pop",
    hover_name="country",
)

fig.show()
````

## III. `fig.update_traces()`


`fig.update_traces()` changes the traces: the visible marks inside the plot.

Use it for things like marker size, marker outline, line width, bar opacity, text position, and hover format.


````python
df = pd.DataFrame({
    "x": [1, 2, 3, 4, 5],
    "y": [4, 6, 5, 8, 7],
    "group": ["A", "A", "B", "B", "B"]
})

fig = px.scatter(df, x="x", y="y", color="group")

fig.update_traces(
    marker=dict(size=14, line=dict(width=2, color="black")),
    opacity=0.8
)

fig.show()
````

You can also update only certain traces by using `selector`.


````python
fig = px.scatter(df, x="x", y="y", color="group")

fig.update_traces(
    marker=dict(size=16, symbol="diamond"),
    selector=dict(name="A")
)

fig.update_traces(
    marker=dict(size=16, symbol="circle"),
    selector=dict(name="B")
)

fig.show()
````

## IV. `fig.update_layout()`


`fig.update_layout()` changes the overall figure layout: title, axis titles, legend, size, margins, theme, background, and more.


````python
df = pd.DataFrame({
    "day": ["Mon", "Tue", "Wed", "Thu", "Fri"],
    "sales": [12, 18, 15, 22, 26]
})

fig = px.bar(df, x="day", y="sales", text="sales")

fig.update_traces(
    marker_color="steelblue",
    marker_line_color="black",
    marker_line_width=1,
    textposition="outside"
)

fig.update_layout(
    title="Daily Sales",
    xaxis_title="Day",
    yaxis_title="Sales",
    width=700,
    height=450,
    margin=dict(l=60, r=30, t=70, b=60)
)

fig.show()
````

## V. Axes, Ticks, Limits, and Grid


Axis settings can be changed inside `fig.update_layout()` or with the convenience methods `fig.update_xaxes()` and `fig.update_yaxes()`.

| Element | Method |
| --- | --- |
| Axis title | `fig.update_layout(xaxis_title=..., yaxis_title=...)` |
| Axis range | `fig.update_xaxes(range=[...])`, `fig.update_yaxes(range=[...])` |
| Tick values | `fig.update_xaxes(tickvals=[...])` |
| Tick labels | `fig.update_xaxes(ticktext=[...])` |
| Grid lines | `fig.update_xaxes(showgrid=True)`, `fig.update_yaxes(showgrid=True)` |


````python
df = pd.DataFrame({
    "month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "temperature": [42, 45, 52, 61, 70, 78]
})

fig = px.line(df, x="month", y="temperature", markers=True)

fig.update_traces(
    line=dict(color="firebrick", width=3),
    marker=dict(size=10)
)

fig.update_layout(
    title="Average Temperature",
    xaxis_title="Month",
    yaxis_title="Temperature (F)",
    template="plotly_white"
)

fig.update_yaxes(range=[35, 85], dtick=10, showgrid=True, gridcolor="lightgray")
fig.update_xaxes(showgrid=False)

fig.show()
````

## VI. Hover Labels and Text


One of Plotly's strengths is interactivity. You can decide what appears when the user hovers over a data point by using `hover_name`, `hover_data`, and `hovertemplate`.


````python
df = px.data.gapminder().query("year == 2007 and continent == 'Asia'")

fig = px.scatter(
    df,
    x="gdpPercap",
    y="lifeExp",
    size="pop",
    color="country",
    hover_name="country",
    hover_data={"gdpPercap": ":,.0f", "lifeExp": ":.1f", "pop": ":,"},
    log_x=True
)

fig.update_traces(marker=dict(line=dict(width=1, color="black")))

fig.update_layout(
    title="GDP per Capita and Life Expectancy in Asia, 2007",
    xaxis_title="GDP per Capita",
    yaxis_title="Life Expectancy",
    template="plotly_white",
    showlegend=False
)

fig.show()
````

`hovertemplate` gives you full control over the tooltip. Use Plotly placeholders such as `%{x}`, `%{y}`, and `%{customdata[0]}` to insert values, and use `<extra></extra>` to hide the secondary trace label.

````python
df = px.data.gapminder().query("year == 2007 and continent == 'Asia'")

fig = px.scatter(
    df,
    x="gdpPercap",
    y="lifeExp",
    size="pop",
    color="country",
    custom_data=["country", "pop"],
    log_x=True
)

fig.update_traces(
    hovertemplate=(
        "<b>%{customdata[0]}</b><br>"
        "GDP per capita: $%{x:,.0f}<br>"
        "Life expectancy: %{y:.1f} years<br>"
        "Population: %{customdata[1]:,}<extra></extra>"
    ),
    marker=dict(line=dict(width=1, color="black"))
)

fig.update_layout(
    title="Custom Hover Template",
    xaxis_title="GDP per Capita",
    yaxis_title="Life Expectancy",
    template="plotly_white",
    showlegend=False
)

fig.show()
````

## VII. Save


Plotly figures can be saved as interactive HTML files. This is one of the most useful ways to share Plotly plots because the interactivity is preserved.

````python
fig.update_layout(autosize = True)
fig.write_html("fig.html", include_plotlyjs='cdn', config={'responsive': True, 'displayModeBar': True})
````

- `autosize` and `config` make the size of the html reponsive to the width of the website.
- `include_plotlyjs='cdn'` makes the file size of the html small. Openint this file requires internet connection.
